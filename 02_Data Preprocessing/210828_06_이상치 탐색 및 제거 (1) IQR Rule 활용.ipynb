```python
df = pd.read_csv('./data/glass.csv')
df.head()
```

```python
# 종속변수, 독립변수 분리
X = df.drop(['Type'], axis=1)
Y = df['Type']
```

```python
# 학습 데이터와 평가 데이터 분리
from sklearn.model_selection import train_test_split
Train_X, Test_X, Train_Y, Test_Y = train_test_split(X, Y)
Train_X.shape
```

```python
# Box Plot 구성 내용
# 0.25 0.5 0.75 1 나눈 얘기
def IQR_Rule(val_list): # 피처 한개에 대해 작업
    #IQR 계산
    
    Q1 = np.quantile(val_list, 0.25)
    Q3 = np.quantile(val_list, 0.75)

    IQR = Q3 - Q1
    
    not_outlier_condition = (Q3 + 1.5 * IQR > val_list) & (Q1 - 1.5 * IQR < val_list)
    
    return not_outlier_condition
```

```python
Train_X
```

```python
conditions = Train_X.apply(IQR_Rule) #apply함수 이용하면 해당되는 df의 컬럼별로 나누어서 돌려줌
conditions
```

```python
conditions.sum(axis=1) == len(Train_X.columns)
```

```python
total_cond = conditions.sum(axis=1) == len(Train_X.columns)
```
