## K-최근접 이웃(K-Nearest Neighbor) 알고리즘을 이용한 분류

### K-최근접 이웃(K-Nearest Neighbor) 알고리즘 소개

- 테스트 데이터와 K개 훈련데이터의 y값들을 비교
- 분류와 회귀 모두 활용
    - 분류 : k개 최근접 이웃들의 class들 중 다수결 결과로 class 예측
    - 회귀 : k개 최근접 이웃들이 가지고 있는 값의 평균을 결과값으로 예측

- 비모수방식이며 instance-based 알고리즘
    - 모수방식 : 알려진 확률분포(정규분포, 이항분포 등)를 기반으로 해당 모수를 추정
    - 비모수방식 : 알려진 확률분포를 기반하지 않고 다른 방식으로 해당 모수를 추정

- K 값 : 동점을 막기 위해 대개 홀수로 정함 
    - k가 작으면 이상치 등 노이즈에 민감하게 반응하여 과대적합 
        - k=1인 경우 가장 가까운 이웃과 같은 클래스가 할당됨
    - k가 크면 자료의 패턴을 파악할 수 없어 과소적합
        - k=N인 경우 전체 평균의 개념


### 사이킷런의 K-최근접이웃 클래스m

```python
from sklearn.neighbors import KNeighborsClassifier
```

#### KNeighbor Class 객체 생성
- 훈련(training) : fit(features, label) 메서드
- 평가 : score() 메서드 => 정확도

```python
kn = KNeighborsClassifier()
```

```python
kn.fit(fish_data, fish_target)
```

```python
kn.score(fish_data, fish_target) #학습한 내용과 동일해서 100%가 나온 것 (workfow 3번 과정을 해야함)
```

**정확도(accuracy)**
​
- 정확한 답을 몇개 맞추었는지를 백분율로 나타낸 값
-  $정확히 맞힌 개수 / 전체 데이터수$

### K-최근접 이웃 알고리즘을 이용한 새로운 데이터 분류

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.scatter(30, 600, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

### 새로운 Data를 활용해서 생성한 모델 kn의 성능을 확인

```python
# 예측 함수 : predict(피처)

kn.predict([[30,600], [50,550],[5,20]])
kn.predict([[1,5]])
kn.predict([[20,900]])
```

#### K-neighbors 크기 지정

```python
# fish_data 전체 크기 49
kn49 = KNeighborsClassifier(n_neighbors=49)
```

```python
kn49.fit(fish_data, fish_target)
```

**49개 중 도미가 35개 이므로 어떤 데이터를 넣어도 무조건 도미로 예측**

```python
kn49.score(fish_data, fish_target) # fish 표본에 갯수가 49에서 35개
35/49 
```

### 확인문제

```python
kn = KNeighborsClassifier() # 모델 인스터스 생성 -n_neighbor 기본값 5
kn.fit(fish_data, fish_target) # 학습

for k in range(5, 50):
    # 최근접 이웃 개수 설정
    kn.n_neighbors = k
    # 점수 계산
    score = kn.score(fish_data, fish_target)
    # print('k=%d score=%f' %(k, score))
    # 100% 정확도에 미치지 못하는 이웃 개수 출력
    print(k,score)
```
